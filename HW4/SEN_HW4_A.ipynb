```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

n = 1000
degree_avg_list = []

for p in np.arange(0, 1, 0.001):
  degree_sum = 0

  for i in range(10):
    G = nx.fast_gnp_random_graph(n, p)
    degree_sum += G.degree[0]

  degree_avg = degree_sum / 10
  degree_avg_list.append(degree_avg)

plt.figure(figsize=(8, 6), dpi=120)
plt.plot(np.arange(0, 1, 0.001), degree_avg_list)
plt.xlabel('p')
plt.ylabel('degree')
plt.show()
```
